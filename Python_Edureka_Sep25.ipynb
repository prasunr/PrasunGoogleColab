```python
# Types of Arguments

#1 Required Positional arguments
#-----------------------------------------
# means arguments are required and also the position matters. Can't pass wrong
def funcWithArg(name, year):
  print(f"Hi, {name} you are {year} years old")

funcWithArg("Prasun", 44)
```

```python
#2 Keyword arguments
#----------------------------------------
# HIGHLY PREFERED METHOD
funcWithArg(name="MichaeL", year=1991)
# ALSO ORDER DOES NOT MATTER HERE
funcWithArg(year=2022,name="dj")
```

```python
#3 Default or optional arguments
#-------------------------------------------
def funcWithDefaultArg(name, year=1991):
  print(f"Hi, {name} you are {year} years old")
funcWithDefaultArg("prasun")
```

```python
#4 Variable-length/Arbitary arguments(*args, **kwargs)
#---------------------------------------------
def func4(*names): #Arbitary arguments
  for name in names:
    print(f"Hi, {name}")

func4("prasun", "ac/dc")

#key-arbitary argument
def func5(**employee_details): #key arbitary argument
  print(employee_details)

func5(id=1001,name="praun")
```

```python
# Class

class Student:
    def __init__(self, name, roll_no):
      self.name = name
      self.roll_no = roll_no
      print("Constructor called")

stud1 = Student("prasun", 1001)
print(stud1.name)
print(stud1.roll_no)
```

```python
# Encapsulation
# Access Modifiers/specifiers: public, private and protected
# By default all variables are public
# __<variable> (double underscore before) means it's a private variable
# _<variable> (single underscore before) means it's a protected variable
class Account:
  def __init__(self, owner, balance):
    self.owner = owner       #public attibute
    self.__balance = balance #private attribute

  def deposit(self, amount):
    self.__balance += amount

  def get_balance(self):
    return self.__balance

acc1 = Account("prasun", 10000)
acc1.deposit(500)
print(acc1.get_balance())
#print(acc1.__balance) will give a compile error.
```

```python
# Inheritance
class Animal:
  def speak(self):
    print("Animal makes a sound")

class Dog(Animal):
  def bark(self):
    print("Dog barks")

dog = Dog()
dog.bark()
dog.speak()

```

```python
# Multiple Inheritance
class A:
  def feature1(self):
    print("Feature A")

class B:
  def feature2(self):
    print("Feature B")

class C(A,B):
  pass  # empty statement - we don't have anything to do.

# telling the interpreter that i'm inheriting both the classes A and B.
c = C()
c.feature1()
c.feature2()
```

```python
class A:
  def feature1(self):
    print("Feature A")

class B:
  def feature2(self):
    print("Feature B")

class C(A,B):
  def feature3(self):
    print("Feature C")

obj2 = C()
obj2.feature1()
obj2.feature2()
obj2.feature3()
```

```python
# Exception Handling
# To handle exceptions, we use
# Try, except, finally
import sys

try:
  var4 = int(input("Enter a number4: "))
  var5 = int(input("Enter a number5: "))
  var6 = var4/var5
  print(var6)

except ValueError:
  print("Invalid input")
  print(sys.exc_info())
except ZeroDivisionError:
  print("Zero division error")
  print(sys.exc_info())
finally:
  print("i'm inside Finally block")

#import sys module and call
#sys.exc_info()
```

```python
#Create numpy array
import numpy as np
# as np is alias. Instead of calling numpy. etc use np.
arr = np.array([1,2,3,4,5])
print(arr)
print(type(arr))

# We cannot store hetrogenious datatypes in numpy.
# all datatypes should be same.

```

```python
#.dtype will give you the datatype of a numpy array
print(arr.dtype)
```

```python
#arange(start, stop, step) here start is inclusive, stop is not.
arr = np.arange(1,10,2)
print(arr)
```

```python
#Random array methods .rand() and .randint(start, stop, size or len of array)
arr = np.random.rand(5)
print(arr)
arrInt = np.random.randint(1,100,5)
print(arrInt)
```

```python
#Create 2D array - make sure you have same number of columns in all rows. I.e here all rows should have 3 columns.
arr = np.array([[10,20,30],[40,50,60],[70,80,90]])
print(arr)

#.ndim returns 2 for 2 dimensional array
print(arr.ndim)

#.shape returns rows and columns in parenthesis like this: (rows, columns)
print(arr.shape)

#How to access elements in a 2d array
print(arr[0][1])
```

```python
# Matrix Multiplication - is done using .dot()
arr1 = np.array([[1,2,3],[4,5,6]])
arr2 = np.array([[10,11],[20,21],[30,31]])
print(arr1) # 2X3
print(arr2) # 3X2
print(np.dot(arr1,arr2)) #Output will be 2x2
```

```python
# Reshape() - reshape a matrix
arr = np.array([40,50,60,70,80,90])
# Reshape it to 2x3
arr_new = arr.reshape(3,2)
print(arr_new)

arr_new = arr.reshape(3,2)
print(arr_new)

#IMPORTANT
# we cannot go against the size of the array
# we cannot have something like
arr_new = arr.reshape(3,3) # has 9 elements but we only have 6 elements
print(arr_new)
```

```python
# Transpose
print(arr_new)
print(np.transpose(arr_new))
# Rows become columns, and columns become rows
```

```python
# Broadcasting of Arithmetic Operations
lst = [10,20,30]
print(lst*5)

arr = np.array([10,20,30])
print(arr*5)
```

```python
# Stats Functions
sales = np.array([456,123,789,489, 453])
print("Total Sales:", sales.sum())
print("Min sales: ", sales.min())
print("Max sales: ", sales.max())
print("Mean sales: ", sales.mean())
print("Median sales: ", np.median(sales))
print("Standard Deviation: ", sales.std())
print("Variance: ", sales.var())
```
